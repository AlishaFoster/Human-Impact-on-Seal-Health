## Background and Prior Work


### Introduction

One of the most important animals to the Arctic ecosystem is the ice seal. These cute, chubby mammals come in many species, such as bearded, ringed, spotted, and ribbon seals. Unfortunately for these rotund aquatic animals, there has been an unusual decline in seal health over the past decade from 2011-2018.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Disease, climate change, and harmful human intervention have all impacted the serene icy landscape of the Arctic Circle, specifically in Alaska. Within the shrinking ice flats of the north, a delicate food web takes place that allows its wildlife to survive. Seals are known to occupy underwater lairs and hunt for fish and other invertebrates at the edge of these ice lands. Their hunting season also lines up with many Alaskan oil and gas activities such as ice road construction and seismic/drilling operations.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) These disturbances to the seals' habit threaten their health and safety. Without seals, the entire balance of the Arctic ecosystem could be destroyed in the future.

### Prior Work

Many research projects have explored these impacts to seal health. One project conducted by the Alaska Department of Fish and Game surveyed ringed seal locations in the Beaufort Sea to see if their population was impacted by drilling operations in the area.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Reasons for this study largely revolved around finding the best times and locations to carry out drilling and seismic construction while disturbing the least amount of seals possible. They found 60 breathing holes and 10 lairs under the ice, with the closest structures being within 19.5m of facilities and within an active ice road.Overall, this project revealed the close proximity of seal habitats to human construction and development, showing how their livelihood has become threatened in recent years.

Another research study conducted in Alaska considers declining seal body fat and how that may be attributed to global warming and receding ice near the Bering Sea.<a name="cite_note-2"></a>[<sup>2</sup>](#cite_ref-2) While the study, which considered population and weather data from 2007-18, does not directly connect seal wellbeing to human overfishing, evidence suggested that decreasing seal health is connected to worsening diets in the Alaska region. Further exploration into the fishing of the region could reveal how fish population interacts with seal body weight. "The mothers, maybe, were not finding as much food in the period prior to the birth of their pups, when they were pregnant," author Boveng said.

One last study touches on specific aspects of seal health, such as disease and body chemistry.<a name="cite_note-3"></a>[<sup>3</sup>](#cite_ref-3) Sea ice has been receding in recent years due to a multitude of factors such as climate change, ice drilling, and road construction. This has caused different types of seals to begin living in close proximity to each other. By taking a small sample of bearded, ringed, spotted, and ribbon seals from all across the Alaskan coast, researchers were able to test for diseases such as Brucella, Leptospira, avian influenza, herpesvirus PhHV-1, and morbillivirus. This paper summarizes that since ice seals have typically had low exposure to disease, their increased exposure to other seal populations has thus increased their mortality rate.

## Alaskan Phocid Morphometrics, 1991 to 2015

```python
# imports
import pandas as pd
import numpy as np
import re
```

```python
# expand max output preview length
pd.set_option('display.max_rows', None)   
morpho_df = pd.read_csv('data/morphometrics_data.csv')
morpho_df.columns
```

```python
# read all valid comments for this dataset
# df2[df2["morphometriccomments"].notna()]["morphometriccomments"]

# drop all nan rows, ignoring speno column (which will never be NaN)
morpho_df = morpho_df.dropna(how='all', subset=morpho_df.columns.difference(['speno']))
```

```python
morpho_df = morpho_df.loc[:, morpho_df.isnull().mean() < 0.8]
morpho_df.info()
```

We need the morphometrics data as a time series to spot correlations and trends with the other datasets, so lets find the years that data was recorded by extracting it from the "speno" variables.

```python
# use regex to get all numerical components of a string
def extract_numerical_components(string):
    numerical_components = re.findall(r'\d+', string)
    numerical_components = [int(component) for component in numerical_components]
    return numerical_components

# figure out how many years we have data for
years = []
for index, row in morpho_df.iterrows():
    speno_value = row['speno']
    # only pay attention to the year component
    year = extract_numerical_components(speno_value)[0]
    years.append(year)

print(years)
unique_years = set(years)
print(unique_years)
```

Next, we can organize it by those years and reformat it using groupby! 

```python
morpho_df['year'] = years
# drop speno col
morpho_time = morpho_df.drop(columns=['speno'])
# group by yr
morpho_time = morpho_time.groupby('year').mean().reset_index()
```

```python
# drop blubber columns since they are mostly NaN
morpho_time = morpho_time.drop(columns=['blubberthickness_cm', 'blubberdepthrightlateralhipa', 'blubberdepthdorsalhipa_cm','blubberdepthdorsalaxillarya','blubberdepthlateralaxillarya'])
```

Is the downward slope in seal births statistically significant, at the 99% confidence level?

To answer this question, we wil run a simple linear regression model with `year` as the dependent variable and `pups_born` as the independent variable. Our regression equation will take the form
$$
y = \beta_0 + \beta_1 x
$$
where
- $y$ represents `pups_born`
- $x$ represents `year`
- $\beta_0$ is the y-intercept
- $\beta_1$ is the slope

We test
$$
H_0: \beta_1 = 0 \quad \text{vs} \quad H_1: \beta_1 < 0
$$

```python
births = seal_population['pups_born']
index_list = []
for i in range(len(seal_population)) :
    if not pd.isnull(seal_population['pups_born'][i]):
        index_list.append(i)
years = seal_population['year'][index_list]
births = births.dropna()

slope, intercept, r_value, p_value, std_err = stats.linregress(years, births)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")
```

**Conclusion**

Since the p-value of $\beta_1 = 0.0002 < 0.01 = \alpha$, therefore we reject $H_0$. Thus, it seems that there is a statistically significant downward trend in the birth of seals over the years 2006-2018. This indicates that there may be potential harmul factors affecting the seal population, which we will further investigate in this project.

# Clearly ridge regression works, but our dataset is so small (test set is literally 3 points) that it will most likely not fit well to unseen data
# Try bootstrapping and pca to make our model more robust:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge
from sklearn.utils import resample
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

X = final_merge.drop(columns=['mass_kg'])
y = final_merge['mass_kg']

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_scaled)

ridge_model = Ridge(alpha=1.0)

# Bootstrapp
n_iterations = 1000
n_size = len(X_reduced)
r2_scores = []
mae_scores = []
mse_scores = []
coefs = np.zeros((n_iterations, X.shape[1]))

for i in range(n_iterations):
    X_train, y_train = resample(X_reduced, y, n_samples=n_size, random_state=i)
    
    ridge_model.fit(X_train, y_train)
    
    y_pred_train = ridge_model.predict(X_train)
    r2_scores.append(r2_score(y_train, y_pred_train))
    mae_scores.append(mean_absolute_error(y_train, y_pred_train))
    mse_scores.append(mean_squared_error(y_train, y_pred_train))
    
    coefs[i, :len(ridge_model.coef_)] = ridge_model.coef_

# Average of metrics
mean_r2 = np.mean(r2_scores)
mean_mae = np.mean(mae_scores)
mean_mse = np.mean(mse_scores)
std_r2 = np.std(r2_scores)
std_mae = np.std(mae_scores)
std_mse = np.std(mse_scores)

print(f"Ridge Regression with Bootstrapping:")
print(f"Mean R2 Score: {mean_r2} Â± {std_r2}")
print(f"Mean MAE: {mean_mae} Â± {std_mae}")
print(f"Mean MSE: {mean_mse} Â± {std_mse}")

# Average coefficient importance
mean_coefs = np.mean(coefs, axis=0)

# Map them back to the original features
important_features = pd.Series(mean_coefs, index=X.columns).sort_values(ascending=False)

print("Most influential features in predicting mass_kg:")
print(important_features.head(10))

plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.hist(r2_scores, bins=30, color='blue', alpha=0.7)
plt.xlabel('RÂ² Score')
plt.ylabel('Frequency')
plt.title('Distribution of RÂ² Scores')

plt.subplot(2, 2, 2)
plt.hist(mae_scores, bins=30, color='green', alpha=0.7)
plt.xlabel('Mean Absolute Error')
plt.ylabel('Frequency')
plt.title('Distribution of MAE Scores')

plt.subplot(2, 2, 3)
plt.hist(mse_scores, bins=30, color='red', alpha=0.7)
plt.xlabel('Mean Squared Error')
plt.ylabel('Frequency')
plt.title('Distribution of MSE Scores')

plt.subplot(2, 2, 4)
important_features.head(4).plot(kind='barh', color='purple', alpha=0.7)
plt.xlabel('Coefficient Value')
plt.title('Top 4 Most Influential Features')

plt.tight_layout()
plt.show()
```

#### What we can gather from these models

Bootstrapping and PCA are used in conjunction to enhance the reliability and interpretability of the Ridge Regression model, especially when dealing with a small dataset like this. Here's why we employ these techniques and how to interpret the results:

Bootstrapping:
Increased Robustness: Bootstrapping involves resampling with replacement from the original dataset to create multiple bootstrap samples. Each sample is used to train a separate Ridge Regression model. By aggregating results across these models, we get a more robust estimate of model performance.

Assessment of Model Stability: Bootstrapping allows us to assess the stability of the model by evaluating its performance across multiple samples. The variability in performance metrics (such as RÂ², MAE, and MSE) provides insights into the model's stability and generalizability.
